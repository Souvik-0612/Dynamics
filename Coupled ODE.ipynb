# Coupled Diffarential equations

Solve the DE using appropiate inital conditons.

$$\frac{dx}{dt}=\sigma (y-x)$$

$$\frac{dy}{dt}=x(\rho - z)-y$$

$$\frac{dz}{dt}=xy - \beta z$$

> This is famous **Lorenz equations**. Here $\sigma$, $\rho$, $\beta$ are parameters: $\sigma$ is the *Prandtl number*, $\rho$ is the Rayleigh number and $\beta$ has no name.

### Coupled 1st Order ODE

```python
def df1(x, y, z): return 10*(y - x)
def df2(x, y, z): return x*(28 - z)- y
def df3(x, y, z): return x*y - 8/3*z

#Intiial condtions
x, y, z = 1, 3, 5
t = 0

dt = 0.01
tf = 20

xx, yy, zz, tt = [], [], [], []
while(t <= tf):
    t += dt
    x += df1(x, y, z)*dt
    y += df2(x, y, z)*dt
    z += df3(x, y, z)*dt
    xx.append(x)
    yy.append(y)
    zz.append(z)
    tt.append(t)
    
#Plotting
plt.rcParams["figure.figsize"] = (18, 14)
plt.subplot(2, 3, 1)
plt.plot(tt, xx, "Brown")
plt.ylabel(r"$x$", size = 20)
plt.xlabel(r'$t$', size = 20)
plt.subplot(2, 3, 2)
plt.plot(tt, yy, "Black")
plt.ylabel(r"$y$", size = 20)
plt.xlabel(r'$t$', size = 20)
plt.subplot(2, 3, 3)
plt.plot(tt, zz, "m")
plt.ylabel(r"$z$", size = 20)
plt.xlabel(r'$t$', size = 20)
plt.subplot(2, 3, 4)
plt.plot(xx, zz, "green")
plt.ylabel(r"$z$", size = 20)
plt.xlabel(r'$x$', size = 20)
plt.subplot(2, 3, 5)
plt.plot(xx, yy, "red")
plt.ylabel(r"$y$", size = 20)
plt.xlabel(r'$x$', size = 20)
plt.subplot(2, 3, 6)
plt.plot(yy, zz, "blue")
plt.ylabel(r"$z$", size = 20)
plt.xlabel(r'$y$', size = 20)
plt.show()
```

### Coupled 2nd Order ODE

```python
def df1(x, y, z): return 10*(y - x)
def df2(x, y, z): return x*(28 - z)- y
def df3(x, y, z): return x*y - 8/3*z

#Intiial condtions
x, y, z = 1, 3, 5
t = 0

dt = 0.01
tf = 20

xx, yy, zz, tt = [], [], [], []
while(t <= tf):
    t += dt
    x += df1(x, y, z)*dt
    y += df2(x, y, z)*dt
    z += df3(x, y, z)*dt
    xx.append(x)
    yy.append(y)
    zz.append(z)
    tt.append(t)
    
#Plotting
plt.rcParams["figure.figsize"] = (18, 14)
plt.subplot(2, 3, 1)
plt.plot(tt, xx, "Brown")
plt.ylabel(r"$x$", size = 20)
plt.xlabel(r'$t$', size = 20)
plt.subplot(2, 3, 2)
plt.plot(tt, yy, "Black")
plt.ylabel(r"$y$", size = 20)
plt.xlabel(r'$t$', size = 20)
plt.subplot(2, 3, 3)
plt.plot(tt, zz, "m")
plt.ylabel(r"$z$", size = 20)
plt.xlabel(r'$t$', size = 20)
plt.subplot(2, 3, 4)
plt.plot(xx, zz, "green")
plt.ylabel(r"$z$", size = 20)
plt.xlabel(r'$x$', size = 20)
plt.subplot(2, 3, 5)
plt.plot(xx, yy, "red")
plt.ylabel(r"$y$", size = 20)
plt.xlabel(r'$x$', size = 20)
plt.subplot(2, 3, 6)
plt.plot(yy, zz, "blue")
plt.ylabel(r"$z$", size = 20)
plt.xlabel(r'$y$', size = 20)
plt.show()
```

To solve the above 2nd Order ODE we need to solve four 1st Order ODE. So recap it. if we have a ODE like this, 
### 2nd Order ODE
$$\frac{d^2y}{dx^2}+\lambda \frac{dy}{dx}+ky=0$$
> Slit the eq: $\frac{dy}{dx}=z$  & $\frac{dz}{dx}=-\lambda z - ky$

##### Euler method: $f(x_n+h) = f(x_n)+hf'(x_n)$

#### The Foucault pendulum

The non-linear diffarential equations become :

$$\ddot x + \frac{gx}{l} - 2\omega \dot y cos(\lambda) - \frac{2\omega x \dot y sin(\lambda)}{l} = 0$$ 
> $\frac{dx}{dt}= z_1$ &  $\frac{dz_1}{dt}=-\frac{gx}{l} + 2\omega \dot y cos(\lambda)+ \frac{2\omega x z_2 sin(\lambda)}{l}$

$$\ddot y + \frac{gy}{l} + 2\omega \dot x cos(\lambda)-\frac{2\omega y \dot y sin(\lambda)}{l}=0$$
> $\frac{dy}{dt}= z_2$ &  $\frac{dz_2}{dt}=-\frac{gy}{l} - 2\omega \dot x cos(\lambda)+ \frac{2\omega y z_2 sin(\lambda)}{l}$

Solve this ODE using initial conditions: at $t=0$, $x = 0$, $\dot x = 0$, $y = A$, $\dot y = 0$

```python
#Constants:
g, l, omg, lam = 10, 100, 2e-7, 0.4

#Initial condtions:
t, x, z1, y, z2 = 0, 0, 0, 10, 0

#Diffraential eq:
def f1z(z1): return z1
def f1(t, x, z1, z2): return -g*x/l + 2*omg*z2*np.cos(lam) + 2*omg*x*z2*np.sin(lam)/l

def f2z(z2): return z2
def f2(t, y, z2, z1): return -g*y/l - 2*omg*z1*np.cos(lam) + 2*omg*y*z2*np.sin(lam)/l

##
tf = 80
h = 0.01

#Empty list
xx, yy, tt = [], [], []

while (tf >= t):
    xx.append(x)
    yy.append(y)
    tt.append(t)
    x += h*z1
    z1 += h*f1(t, x, z1, z2)
    y += h*z2
    z2 += h*f2(t, y, z2, z1)
    t += h  
    
plt.plot(xx, yy)
plt.show()
```

### Analytical solutions
using lots of assumptions

$$x = A cos(\sqrt{g/l} t)sin(\omega cos(\lambda)t)$$
$$y = A cos(\sqrt{g/l} t)cos(\omega cos(\lambda)t)$$

```python
#Constants:
A, g, l, ome, lam = 10, 10, 100, 2e-7, 0

t = np.arange(0, 60, 0.01)
x = A*np.cos(np.sqrt(g/l)*t)*np.sin(ome*np.cos(lam)*t)
y = A*np.cos(np.sqrt(g/l)*t)*np.cos(ome*np.cos(lam)*t)

plt.plot(x, y, "red")
plt.show()
```

# Elicptical orbit 
*Newton's law of gravition*

$$\vec F = - \frac{GMm}{r^3} \vec r$$


$$\frac{d^2 \vec r}{dt^2} = - \frac{GM}{r^3} \vec r$$



#### So the equation of motion in cartesian plane are $\frac{d^2 x}{dt^2} = - \frac{GM}{r^3} x$ and $\frac{d^2 y}{dt^2} = - \frac{GM}{r^3} y$ where $r = \sqrt {x^2+y^2}$

I am going to same initial conditions as in the Feynman. At $t=0$, $x = .500$, $y = 0.000$, $v_x = 0.000$, $v_y=1.630$.

```python
#Initial conditions 
t, x, vx, y, vy = 0, 0.5, 0, 0, 1.63

## 
tf = 2.5
h = 0.001 

#Constants:
G = 1
M = 1

r = pow((x**2 + y**2), 0.5)
def f1(x, y): return - G*M*x/r**3
def f2(x, y): return - G*M*y/r**3

xx, yy, tt = [], [], []

while (t <= tf):
    tt.append(t)
    xx.append(x)
    yy.append(y)
    x += h*vx
    vx += h*f1(x, y)
    y += h*vy
    vy += h*f2(x, y)
    t += h
    
plt.figure(figsize = (10, 10))
plt.plot(xx, yy)
plt.show()
```
